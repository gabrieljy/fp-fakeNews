```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi

# Define categories and methods
categories = ['Accuracy', 'Scalability', 'Explainability', 'Speed', 'Generalizability']
methods = ['Passive-Aggressive', 'Random Forest', 'BERT', 'ClaimBuster', 'FaKnow', 'Human-operated']

# Define values for each method
values = [
    [6, 5, 6, 8, 3],  # Passive-Aggressive
    [7, 6, 7, 7.1, 6],  # Random Forest
    [10, 9, 4, 7.2, 10], # BERT
    [8, 7, 7, 7.3, 6],  # ClaimBuster
    [9, 7, 8, 6.9, 8],  # FaKnow
    [10, 1, 10, 2, 10] # Human-operated
]

# Normalize data and prepare for plotting
angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]  # Close the circle

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Plot each method
for i, method in enumerate(methods):
    data = values[i] + [values[i][0]]  # Close the circle for each method
    ax.plot(angles, data, linewidth=1, linestyle='solid', label=method)
    ax.fill(angles, data, alpha=0.25)

# Add category labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, fontsize=10)

# Add a title and legend
plt.title('Comparison of Fake News Detection Methods', size=15, y=1.1)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1), fontsize=9)

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch

# Define layers and their positions
layers = {
    "Input Layer": (0, 0),
    "Data Layer": (1, 1),
    "Modeling Layer": (2, 2),
    "Application Layer": (3, 1),
    "Evaluation & Monitoring Layer": (4, 2),
    "Deployment Layer": (5, 0)
}

# Define connections
connections = [
    ("Input Layer", "Data Layer"),
    ("Data Layer", "Modeling Layer"),
    ("Modeling Layer", "Application Layer"),
    ("Application Layer", "Evaluation & Monitoring Layer"),
    ("Evaluation & Monitoring Layer", "Deployment Layer"),
]

# Adjust box size dynamically
def text_box_size(text, scale=0.05):
    return len(text) * scale * 1.5, 0.5  # Width based on text length, height constant

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(-1, 6)
ax.set_ylim(-1, 3)
ax.axis('off')

# Plot each layer
for layer, pos in layers.items():
    box_width, box_height = text_box_size(layer)
    box = FancyBboxPatch(
        (pos[0] - box_width / 2, pos[1] - box_height / 2),
        box_width,
        box_height,
        boxstyle="round,pad=0.1",
        edgecolor="black",
        facecolor="#ADD8E6"
    )
    ax.add_patch(box)
    ax.text(pos[0], pos[1], layer, ha="center", va="center", fontsize=10, color="black")

# Draw connections
for start, end in connections:
    start_pos = layers[start]
    end_pos = layers[end]
    ax.annotate(
        "",
        xy=(end_pos[0], end_pos[1]),
        xytext=(start_pos[0], start_pos[1]),
        arrowprops=dict(arrowstyle="-|>", color="black")
    )

plt.title("System Architecture Diagram", fontsize=14)
plt.show()
```
